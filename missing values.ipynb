We'll apply our new skills to a dataset of building permits issued in San Francisco.

```python
#module we need
import numpy as np
import pandas as pd

#reading all the data
Building_sf= pd.read_csv("C:\\Users\\hp\\Desktop\\data handling\\archive\\Building_Permits.csv") 


np.random.seed(0)
    
```

```python
# look at the first five rows of the Building_sf file. 
# I can see a handful of missing data already!
Building_sf.head()
```

```python
Building_sf.info()
```

Yep, it looks like there's some missing values.

```python
#get the number of missing datapoints as colomn
missing_values_count= Building_sf.isnull().sum()
missing_values_count[0 : 41]
```

That seems like a lot! It might be helpful to see what percentage of the values in our dataset were missing to give us a better sense of the scale of this problem:

```python
#how many total missing value we have
total_cells = np.product(Building_sf.shape)
total_missing = missing_values_count.sum()
#precent of data that is missing
percent_missingdata=(total_missing/total_cells)*100
print(percent_missingdata)    
```

Wow, almost a quarter of the cells in this dataset are empty! In the next step, we're going to take a closer look at some of the columns with missing values and try to figure out what might be going on with them.

Drop missing values

```python
# remove all the rows that contain a missing value
Building_sf.dropna()
```

Oh dear, it looks like that's removed all our data! ðŸ˜± This is because every row in our dataset had at least one missing value. We might have better luck removing all the columns that have at least one missing value instead.

```python
# remove all columns with at least one missing value
columns_with_na_dropped = Building_sf.dropna(axis=1)
columns_with_na_dropped.head()
```

```python
# just how much data did we lose?
print("Columns in original dataset: %d \n" % Building_sf.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```

We've lost quite a bit of data, but at this point we have successfully removed all the NaN's from our data.
